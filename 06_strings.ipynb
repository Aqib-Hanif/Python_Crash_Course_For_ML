## Creating Strings

Strings are created using:
- Single quotes `' '`  
- Double quotes `" "`  

```python
# Creating strings
name = "Aqib"
message = 'Hello World'

print(name)
print(message)
```

**Explanation:**

- Strings are enclosed in quotes  
- Both single and double quotes work the same  
- `print()` displays the string

## Accessing Characters

Each character in a string has an **index (position)**.

- Index starts from **0**  

```python
# Accessing characters
text = "Python"

print(text[0])  # First character
print(text[3])  # Fourth character
```

**Explanation:**

- `text[0]` → returns 'P'  
- `text[3]` → returns 'h'  
- Index starts from 0

## String Slicing

Slicing allows us to get a **part of a string**.

Syntax: `string[start:end]`

```python
# String slicing
text = "Python"

print(text[0:3])  # Pyt
print(text[2:6])  # thon
```

**Explanation:**

- `text[0:3]` → characters from index 0 to 2  
- `text[2:6]` → characters from index 2 to 5  
- End index is **not included**

## String Methods

Python provides built-in methods to work with strings:
- `upper()` → converts to uppercase  
- `lower()` → converts to lowercase  
- `strip()` → removes spaces  
- `replace()` → replaces text  

```python
# String methods
text = "  hello world  "

print(text.upper())
print(text.lower())
print(text.strip())
print(text.replace("world", "Python"))
```

**Explanation:**

- `upper()` → HELLO WORLD  
- `lower()` → hello world  
- `strip()` → removes spaces from start and end  
- `replace()` → replaces 'world' with 'Python'

## String Formatting

String formatting allows us to **insert values into strings**.

```python
# String formatting
name = "Aqib"
age = 21

print(f"My name is {name} and I am {age} years old")
```

**Explanation:**

- `f""` → f-string formatting  
- `{name}` and `{age}` → insert variables into string  
- Makes output more readable and dynamic
