In this lesson, we will learn about **lists in Python**.

Lists are used to store **multiple values in a single variable**.  
They are very important in **data handling and Machine Learning**, where we work with collections of data.

## Creating Lists

A list is created using **square brackets `[]`**.

Lists can store:
- Numbers  
- Strings  
- Mixed data types  

```python
# Creating lists
numbers = [1, 2, 3, 4, 5]
names = ["Ali", "Ahmed", "Sara"]
mixed = [1, "Hello", 3.5]

print(numbers)
print(names)
print(mixed)
```

**Explanation:**

- Lists are written inside `[]`  
- `numbers` contains integers  
- `names` contains strings  
- `mixed` contains different data types  
- `print()` displays the list

## Accessing List Elements

Each item in a list has an **index (position)**.

- Index starts from **0**
- First element → index 0  
- Second element → index 1  

```python
# Accessing elements
numbers = [10, 20, 30, 40]

print(numbers[0])  # First element
print(numbers[2])  # Third element
```

**Explanation:**

- `numbers[0]` → returns first element → 10  
- `numbers[2]` → returns third element → 30  
- Index always starts from 0

## Modifying Lists

Lists are **mutable**, which means we can change their values.

```python
# Modifying list
numbers = [10, 20, 30]

numbers[1] = 25  # Change second element

print(numbers)
```

**Explanation:**

- `numbers[1] = 25` → changes value at index 1  
- Original: `[10, 20, 30]`  
- Updated: `[10, 25, 30]`

## Adding Elements

We can add elements using:
- `append()` → adds at the end  
- `insert()` → adds at a specific position  

```python
# Adding elements
numbers = [1, 2, 3]

numbers.append(4)      # Add at end
numbers.insert(1, 10)  # Add at index 1

print(numbers)
```

**Explanation:**

- `append(4)` → adds 4 at the end  
- `insert(1, 10)` → adds 10 at index 1  
- Result: `[1, 10, 2, 3, 4]`

## Removing Elements

We can remove elements using:
- `remove()` → removes specific value  
- `pop()` → removes by index  

```python
# Removing elements
numbers = [1, 2, 3, 4]

numbers.remove(2)   # Remove value 2
numbers.pop(1)      # Remove element at index 1

print(numbers)
```

**Explanation:**

- `remove(2)` → removes value 2  
- `pop(1)` → removes element at index 1  
- Result: `[1, 4]`

## Iterating Over Lists

We use loops to go through each element in a list.

```python
# Iterating through a list
numbers = [1, 2, 3, 4]

for num in numbers:
    print(num)
```

**Explanation:**

- `for num in numbers` → goes through each element  
- Each value is stored in `num`  
- `print(num)` → prints each element one by one
