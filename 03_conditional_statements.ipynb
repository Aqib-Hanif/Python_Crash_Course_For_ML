# **Conditional Statements**

In this lesson, we will learn **conditional statements** in Python.

Conditional statements allow programs to **make decisions** based on certain conditions.  
They are very important in **control flow** and **Machine Learning**, like deciding which actions to take depending on data.

## The `if` Statement

The `if` statement checks a **condition**.  
If the condition is **True**, the code inside the `if` block runs.  

Example operators to use inside `if`:
- Comparison operators: `==, !=, >, <, >=, <=`
- Logical operators: `and, or, not`

```python
# If statement example
a = 10

if a > 5:
    print("a is greater than 5")
```

## The `if-else` Statement

The `if-else` statement runs one block if the condition is **True**,  
and another block if the condition is **False**.

```python
# If-Else example
a = 3

if a > 5:
    print("a is greater than 5")
else:
    print("a is not greater than 5")
```

**Explanation:**

- Condition: `a > 5` → False because `a = 3`  
- The `else` block runs → output: "a is not greater than 5"

```python
# Elif example
a = 7

if a > 10:
    print("a is greater than 10")
elif a > 5:
    print("a is greater than 5 but less than or equal to 10")
else:
    print("a is 5 or less")
```

**Explanation:**

1. `a > 10` → False  
2. `a > 5` → True → this block runs → output: "a is greater than 5 but less than or equal to 10"  
3. The `else` block is ignored because one condition was True

## Logical Operators: `and` / `or`

Logical operators let us **combine multiple conditions** inside `if`, `elif`, or `else`.  

- `and` → True if **both conditions** are True  
- `or`  → True if **at least one condition** is True  

These are often used to make **complex decisions** in programs.

```python
# Logical operators example
age = 20
marks = 85

# Using 'and'
if age >= 18 and marks >= 80:
    print("Eligible for scholarship")  # Runs because both are True

# Using 'or'
if age < 18 or marks < 50:
    print("Not eligible for exam")    # Does NOT run because both are False
else:
    print("Eligible for exam")        # Runs instead
```

**Explanation:**

1. `age >= 18 and marks >= 80` → Both True → prints "Eligible for scholarship"  
2. `age < 18 or marks < 50` → Both False → `or` condition is False → else block runs → prints "Eligible for exam"

**Tip:**  
- Use `and` when **all conditions must be True**  
- Use `or` when **any one condition is enough**  
- Combine these with `if`, `elif`, `else` for **flexible decision-making**
