## Comparison Operators

Comparison operators let us **compare two values**.  
Python returns **True** or **False**.

Common operators:
- `==` → equal to  
- `!=` → not equal to  
- `>`  → greater than  
- `<`  → less than  
- `>=` → greater than or equal to  
- `<=` → less than or equal to

```python
# Comparison examples
a = 10
b = 5

print("a == b:", a == b)   # False
print("a != b:", a != b)   # True
print("a > b:", a > b)     # True
print("a < b:", a < b)     # False
print("a >= b:", a >= b)   # True
print("a <= b:", a <= b)   # False
```

**Explanation:**

- `a == b` → checks if a is equal to b → **False**
- `a != b` → checks if a is not equal to b → **True**  
- `a > b` → checks if a is greater than b → **True** 
- `a < b` → checks if a is less than b → **False**
- `a >= b` → checks if a is greater than or equal to b → **True**  
- `a <= b` → checks if a is less than or equal to b → **False**

## Logical Operators

Logical operators let us **combine multiple conditions**.  
Python returns **True** or **False**.

Common logical operators:
- `and` → True if **both conditions** are True  
- `or`  → True if **at least one condition** is True  
- `not` → **reverses** True/False

```python
# Logical examples
x = 10
y = 5
z = 7

print("x > y and x > z:", x > y and x > z)  # True
print("x > y or x < z:", x > y or x < z)    # True
print("not(x > y):", not(x > y))            # False
```

**Explanation:**

1. `x > y and x > z` → True if both conditions are True → **True**
2. `x > y or x < z` → True if at least one condition is True → **True** 
3. `not(x > y)` → reverses the result of `x > y` → **False**
