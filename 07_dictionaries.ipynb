In this lesson, we will learn about **dictionaries in Python**.

Dictionaries are used to store data in **key-value pairs**.  
They are very useful when working with **structured data**, especially in **Machine Learning and data analysis**.

## Creating Dictionaries

A dictionary is created using **curly braces `{}`**.

Each item has:
- a **key**
- a **value**

```python
# Creating a dictionary
student = {
    "name": "Aqib",
    "age": 21,
    "course": "ML"
}

print(student)
```

**Explanation:**

- `"name"`, `"age"`, `"course"` → keys  
- `"Aqib"`, `21`, `"ML"` → values  
- Data is stored in **key-value pairs**

## Accessing Values

We access values using their **keys**.

```python
# Accessing values
student = {
    "name": "Aqib",
    "age": 21
}

print(student["name"])
print(student["age"])
```

**Explanation:**

- `student["name"]` → returns "Aqib"  
- `student["age"]` → returns 21  
- Keys are used instead of index

## Updating Dictionary

We can **modify or add values** using keys.

```python
# Updating dictionary
student = {
    "name": "Aqib",
    "age": 21
}

student["age"] = 22        # Modify value
student["city"] = "Peshawar"  # Add new key-value

print(student)
```

**Explanation:**

- `student["age"] = 22` → updates existing value  
- `student["city"] = "Peshawar"` → adds new key-value pair  
- Dictionaries are **mutable**

## Removing Elements

We can remove items using:
- `pop()` → removes by key  

```python
# Removing elements
student = {
    "name": "Aqib",
    "age": 21,
    "city": "Peshawar"
}

student.pop("age")

print(student)
```

**Explanation:**

- `pop("age")` → removes key "age" and its value  
- Result: `{"name": "Aqib", "city": "Peshawar"}`

## Iterating Over Dictionary

We can loop through:
- Keys  
- Values  
- Key-value pairs  

```python
# Iterating dictionary
student = {
    "name": "Aqib",
    "age": 21,
    "city": "Peshawar"
}

# Keys
for key in student:
    print(key)

# Values
for value in student.values():
    print(value)

# Key-value pairs
for key, value in student.items():
    print(key, ":", value)
```

**Explanation:**

- `for key in student` → loops through keys  
- `values()` → returns all values  
- `items()` → returns key-value pairs  
- Helps in processing data easily
