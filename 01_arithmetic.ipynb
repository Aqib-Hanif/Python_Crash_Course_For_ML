## Addition

Addition is used to **add two numbers together**.

In Python, the **`+` operator** is used for addition.

```python
# Addition example
a = 10
b = 5
print("Addition:", a + b)
```

**Explanation:**

- `a = 10` → stores the number 10 in variable `a`  
- `b = 5` → stores the number 5 in variable `b`  
- `a + b` → adds the two numbers together  
- `print()` → displays the result on the screen  
- Output: `15`

## Subtraction

Subtraction is used to **find the difference between two numbers**.

In Python, we use the **`-` operator**.

```python
# Subtraction example
a = 10
b = 5
print("Subtraction:", a - b)
```

**Explanation:**

- `a - b` → subtracts `b` from `a`  
- Output: `5`  
- The `print()` function displays the result

## Multiplication

Multiplication is used to **multiply two numbers**.

In Python, we use the **`*` operator**.

```python
# Multiplication example
a = 10
b = 5
print("Multiplication:", a * b)
```

**Explanation:**

- `a * b` → multiplies `a` and `b`  
- Output: `50`  
- Multiplication is used in calculations and scaling in ML

## Division

Division is used to **divide one number by another**.

In Python, we use the **`/` operator**, which always returns a **decimal (float)**.

```python
# Division example
a = 10
b = 5
print("Division:", a / b)
```

**Explanation:**

- `a / b` → divides 10 by 5  
- Output: `2.0` (Python always returns a float for division)  
- Division is used in ML for **normalizing data** and calculations

## Floor Division

Floor division **removes the decimal part** and gives only the integer result.

In Python, we use the **`//` operator**.

```python
# Floor division example
a = 10
b = 3
print("Floor Division:", a // b)
```

**Explanation:**

- `10 // 3` → divides 10 by 3  
- Output: `3` (decimal part removed)  
- Useful when you only need **whole numbers**

## Modulus (Remainder)

Modulus returns the **remainder after division**.

In Python, we use the **`%` operator**.

```python
# Modulus example
a = 10
b = 3
print("Remainder:", a % b)
```

**Explanation:**

- `10 % 3` → divides 10 by 3 and returns the remainder  
- Output: `1`  
- Modulus is used to check **even/odd numbers**, cycles, and ML indexing

## Exponent (Power)

Exponent raises a number to the power of another.

In Python, we use the **`**` operator**.

```python
# Exponent example
a = 2
b = 3
print("Power:", a ** b)
```

**Explanation:**

- `2 ** 3` → 2 raised to the power of 3  
- Output: `8`  
- Exponent is useful in ML for **polynomial features, scaling, and calculations**

## BODMAS / Order of Operations

Python follows **BODMAS rules** for calculations:

- **B** → Brackets `( )`  
- **O** → Order / Exponents `**`  
- **D** → Division `/`  
- **M** → Multiplication `*`  
- **A** → Addition `+`  
- **S** → Subtraction `-`  

This means **operations inside brackets are done first**, then exponents, then multiplication/division, and finally addition/subtraction.

```python
# BODMAS example

# Example 1
result1 = 2 + 3 * 4
print("2 + 3 * 4 =", result1)  # Multiplication done first → 2 + 12 = 14

# Example 2
result2 = (2 + 3) * 4
print("(2 + 3) * 4 =", result2)  # Brackets done first → 5 * 4 = 20

# Example 3
result3 = 10 - 2 ** 2
print("10 - 2 ** 2 =", result3)  # Exponent first → 10 - 4 = 6
```

**Explanation:**

1. `2 + 3 * 4`  
   - Multiplication first: `3 * 4 = 12`  
   - Then addition: `2 + 12 = 14`  

2. `(2 + 3) * 4`  
   - Brackets first: `2 + 3 = 5`  
   - Then multiplication: `5 * 4 = 20`  

3. `10 - 2 ** 2`  
   - Exponent first: `2 ** 2 = 4`  
   - Then subtraction: `10 - 4 = 6`  

**Tip:** Always use brackets to make the calculation order **clear and readable**.
