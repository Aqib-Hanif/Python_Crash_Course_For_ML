# **Tuples in Python**

Welcome to the fifth lesson! 🎉

In this lesson, we will learn about **tuples in Python**.

Tuples are used to store **multiple values in a single variable**, just like lists.  
But unlike lists, tuples are **immutable**, meaning their values **cannot be changed** after creation.

## Creating Tuples

A tuple is created using **parentheses `()`**.

Tuples can store:
- Numbers  
- Strings  
- Mixed data types  

```python
# Creating tuples
numbers = (1, 2, 3, 4)
names = ("Ali", "Ahmed", "Sara")
mixed = (1, "Hello", 3.5)

print(numbers)
print(names)
print(mixed)
```

**Explanation:**

- Tuples are written inside `()`  
- `numbers` contains integers  
- `names` contains strings  
- `mixed` contains different data types  
- `print()` displays the tuple

## Accessing Tuple Elements

Each item in a tuple has an **index (position)**.

- Index starts from **0**  
- First element → index 0  

```python
# Accessing elements
numbers = (10, 20, 30, 40)

print(numbers[0])  # First element
print(numbers[2])  # Third element
```

**Explanation:**

- `numbers[0]` → returns first element → 10  
- `numbers[2]` → returns third element → 30  
- Index starts from 0

## Immutability of Tuples

Tuples are **immutable**, which means:
- You **cannot change** their values  
- You **cannot add or remove** elements  

This makes tuples **safe and stable** for storing fixed data.

```python
# Immutability example
numbers = (10, 20, 30)

# This will cause an error
# numbers[1] = 25
```

**Explanation:**

- Tuples do not allow changes after creation  
- Trying to modify (`numbers[1] = 25`) will cause an error  
- This ensures data remains unchanged

## Basic Tuple Operations

Although tuples are immutable, we can:
- Find length using `len()`  
- Count values using `count()`  
- Find index using `index()`  

```python
# Tuple operations
numbers = (1, 2, 3, 2, 4)

print("Length:", len(numbers))
print("Count of 2:", numbers.count(2))
print("Index of 3:", numbers.index(3))
```

**Explanation:**

- `len()` → returns total number of elements  
- `count(2)` → counts how many times 2 appears  
- `index(3)` → returns position of value 3

## Tuple vs List

| Feature      | List        | Tuple       |
|-------------|------------|------------|
| Mutable     | Yes        | No         |
| Syntax      | `[]`       | `()`       |
| Performance | Slower     | Faster     |
| Use case    | Changeable data | Fixed data |
